```python
import numpy as np
import os
import pandas as pd
```

```python
# import data
file_path = 'C:/Users/seanj/Documents/GitHub/business_trends/data/business-licences.csv'
biz_origin = pd.read_csv(file_path, sep=';', low_memory=False)

# set to display all col
pd.set_option('display.max_columns', None)
```

```python
biz_origin.info()
```

```python
# filter for only relevant columns
biz = biz_origin.filter(['LicenceRSN','Status','IssuedDate','BusinessType','BusinessSubType','PostalCode','LocalArea','geo_point_2d'])

biz.head()
```

```python
# update BusinessSubType for businesses that don't require one
biz.update(biz[['BusinessSubType']].fillna('N/A'))
```

```python
# rename col name
biz = biz.rename(columns={'geo_point_2d':'Coordinates'})
```

```python
# change data type to str
biz['LicenceRSN'] = biz['LicenceRSN'].astype(str)
```

```python
# extract year, month from IssuedDate
biz['Year'] = biz['IssuedDate'].str[:4]
biz['Month'] = biz['IssuedDate'].str[5:7]
```

```python
# drop IssuedDate
biz = biz.drop(columns=['IssuedDate'])

# reindex col order
col_order = ['LicenceRSN','Status','Year','Month','BusinessType','BusinessSubType','PostalCode','LocalArea','Coordinates']

biz = biz.reindex(columns=col_order)
```

```python
biz.head()
```

```python
# export as csv
biz.to_csv('clean_biz.csv',index=False)
```
